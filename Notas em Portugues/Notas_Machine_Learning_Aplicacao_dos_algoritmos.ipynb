##Naive-Bayes
O algoritmo de Naive-Bayes é, talvez, um dos mais simples algoritmos de aprendizado de máquina existentes hoje em dia, este é baseado no teorema das probabilidades condicionais postulado por Thomas Bayes. O termo *naive* vem da "ingenuidade" de considerar que as variáveis presentes na base de dados não possuem qualquer correlação. Algo que não se mostra  verdadeiro para a grande maioria dos casos. Porém, este algoritmo pode ser muito interessante de ser aplicado para casos em que a base de dados possui um volume relativamente pequeno e suas variáveis não possuem fortes correlações entre si.

Para a aplicação do algoritmo em python, utilizaremos a biblioteca sklearn, importando o método a seguir

```python
from sklearn.naive_bayes import GaussianNB
```

####Curva ROC - Naive Bayes: Base de crédito

A curva ROC define o quão bom um classificador é em separar os dados em seus respectivos grupos. Por isso, a curva é realizada tendo a taxa de verdadeiros positivos (TPR) em função da taxa de falsos positivos (FPR), realizada, geralmente, para valores diferentes de threshold. O threshold é o limite estipulado para a classificação de um objeto como pertecente a um grupo ou outro, no exemplo da regressão logística, se um dado de entrada tem a probabilidade de pertencer ao grupo A estipulada em >50%, então este é classificado como pertencente a este grupo. Diferentes thresholds são utilizados para diferentes classificadores e a escolha do mais adequado para uma determinada base de dados pode ser melhor feita ao se analisar a curva ROC.

Outros ponto que eu gostaria de abordar na análise da curva ROC é sobre a sua interpretação assintótica, e como ela influencia na forma  como veremos o classificador. Para isso, temos dois casos, o do classificador ideal e o do pior classificador possível.

image.png

Nesse caso, temos o classificador ideal, ou o classificador que consegue separar 100% de todos os dados nos seus respectivos grupos. Ou seja, acertar 100% dos positivos e 100% dos negativos, ou seja, 100% de taxa de verdadeiros positivos e 0% de taxa de falsos positivos. Isso representaria o caso em que todos os dados foram classificados como positivos.

image.png

Aqui, temos o pior caso possível, ou o classificador que classifica todos os reais positivos como negativos, e todos os reais negativos como positivos.

image.png

No caso acima, temos o classificador aleatório.

##Árvore de Decisão
Prosseguimos com a nossa escalada pelos algoritmos de aprendizado de máquina, agora tomando como estudo o algoritmo de árvore de decisão. 
A árvore de decisão tem por objetivo simplificar a rotulação dos dados ao subdividí-los, em termos de seus atributos, em conjuntos menores. Em geral, essa sudbdivisão ocorre na forma de uma árvore, onde os nós que ligam os galhos possuem uma pergunta booleana, e a combinação das respostas para estas perguntas levam o dado para um determinado subconjunto para ser rotulado, como podemos ver no exemplo abaixo, podemos usar o algoritmo de árvore de decisão, para decidir ir à praia ou não.


whatsapp-image-2017-10-29-at-22-27-26.jpeg

Como visto no exemplo acima, as árovres de decisões permitem, a partir de uma série de classificações do dado, facilitar a rotulação do mesmo.
Algumas das vantagens das árvores de decisão estão na boa performance com banco de dados de treinamento pequenos, de funcionar tanto com dados numéricos como categóricos, e também de fácil visualização, em contrapartida, é um algoritmo que pode sofrer bastante com o problema de overfitting.

##Aprendizagem por instância - kNN
Aqui iniciamos o estudo de uma nova estratégia de aprendizagem de máquina, que é a aprendizagem por instância, ou kNN. Este algoritmo busca classificar uma amostra a partir da proximidade desta com as k amostras mais próximas. Por exemplo, imagine que há o grupo A e o grupo B, se desejamos classificar uma amostra que está mais próxima dos elementos do grupo B, então provavelmente ela fará parte também do grupo B. imagem_2021-09-28_233059.png

Porém como tudo na vida, isso depende. Isto é, porque esta classificação depende da quantidade de pontos que consideramos vizinhos ao dado a ser classificado. Por exemplo, no caso da figura acima, se o critério for os 3 vizinhos mais próximos, vemos que dos 3 pontos mais próximos, 2 correspondem à classe B, enquanto que 1 à classe A, o que resultaria na classificação do dado como pertencente à classe B. Porém se considerarmos como critério de classificação os 7 vizinhos mais próximos, 4 destes estariam na classe A, enquanto 3 estariam na classe B, portanto, por sua vez, o ponto seria classificado como sendo da classe A.

Agora, podemos iniciar a implementação do algoritmo nas bases de dados de crédito e censo. Para isso, importamos o classificador na linha abaixo

```python
from sklearn.neighbors import KNeighborsClassifier
```

###kNN - Base de dados de crédito: 98,6%

Agora podemos definir o classificador para esta base de dados, que irá ficar como

```python
knn_credit = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
```

No classificador definimos 3 parâmetros, o primeiro, `n_neighbors` define o número de vizinhos próximos considerados para a classificação, a métrica de minkowski é a função utilizada para calcular a distância entre dois pontos, sendo que p=1, identifica a distância de manhattan, e p=2, identifica a distância euclideana na forma da métrica.

Sabendo disso, realizamos o treinamento do classificador

```python
knn_credit.fit(x_credit_treinamento, y_credit_treinamento)
```

E agora, podemos calcular as previsões, sendo

```python
knn_previsoes_credit = knn_credit.predict(x_credit_teste)
```

E agora, calculamos a precisão do modelo utilizando o `accuracy_score`

```python
accuracy_score(y_credit_teste, knn_previsoes_credit)
```

Obtemos, portanto, uma precisão de 98,6%, um valor muito maior do que encontrado em qualquer outro algoritmo utilizado até agora. Obtendo a matriz de confusão e o relatório de classificação, podemos mostrar as últimas métricas de precisão do classificador.

```python
confusion_matrix(y_credit_teste, knn_previsoes_credit)
```

```python
print(classification_report(y_credit_teste, knn_previsoes_credit))
```

##Regressão Logística

A regressão logística é muito baseada na ideia regressão linear, porém redesenhada para ser aplicada em casos em que a variável dependente (variável alvo) é categórica. Isso acontece em paralelo a um algoritmo de regressão linear, onde o produto da regressão linear é aplicado a uma função sigmoide, onde este resultado é utilizado para comparação com a base de teste. Ou seja, em vez de compararmos os pesos com o resultado da regressão linear, a comparação é feita com o que é retornado pela função sigmoide.
image.png

##Support Vector Machine - SVM
Neste ponto da nota, podemos ter notado que o problema de classificação, na verdade, é um problema geométrico. O que sempre se tenta fazer é construir uma curva que melhor diferencia dois conjuntos de amostra de dados, então valores a partir de um ponto da curva, podem ser classificados como sendo do grupo A, ou B. Como mostrado na imagem a seguir

image.png

Nela, temos a classe positiva e a classe negativa. Caso um novo dado seja inserido na base de dados, o que definirá ele como sendo da classe negativa ou positivo, será a sua posição em relação à reta em verde, chamada de hiperplano de separação. Isso é o que chamamos de problema linearmente separável, pois conseguimos separar o conjunto de dados utilizando apenas uma linha reta.

Mas e quando o conjunto de dados não é linearmente separável? Bem, neste caso, há várias maneiras de se estimar o hiperplano de separação e, inclusive, utilizamos algumas ao longo do texto. Porém, agora, iremos focar em um algoritmo que visa criar um hiperplano de separação ao alterar a dimensão do espaço da base de dados, esse algoritmo é o chamado Support Vector Machine (SVM). Porém, tentarei esclarecer melhor como esse algoritmo realmente funciona.

Como mostrado na imagem acima, a distribuição dos dados no espaço bidimensional, permite que estes sejam separados apenas como uma única linha reta. Porém, e se tivermos uma base de dados, cujos dados estão distribuídos no plano da seguinte maneira,

image.png

Nesse caso, a distribuição dos dados no plano fazem este ser um problema não-linearmente separável. A baixa dimensionalidade do problema, ou o pouco número de parâmetros, pode criar essa não-linearidade na separação dos dados, algo que as vezes é resolvido com a inserção de uma nova característica pros dados, o que aumenta a dimensionalidade do espaço e remodela a distribuição dos dados neste novo espaço, podendo resultado em um problema linearmente separável, como mostrado abaixo

image.png

Mas quase sempre esta estratégia não é viável no mundo real, porém o SVM, utilizando a estratégia de Kernels, consegue simular novos atributos no banco de dados. Basicamente, o truque de kernels consiste na aplicação de funções não-lineares (os kernels) nos dados de entrada do algoritmo, isso tem como consequência um aumento na dimensão do espaço em que os dados estão distribuídos (algo como inserir novos atributos), o que pode tornar o problema linearmente separável. Na prática, geralmente não conseguimos selecionar um kernel que torna um real problema linearmente separável 100%, mas a escolha do kernel correto pode nos ajudar a aprimorar a precisão do nosso modelo.

Com estas informações em mãos, agora podemos iniciar o estudo do algoritmo de SVM, com isso vamos importar seguir com os passos já conhecidos ao longo do nosso texto

##Redes Neurais Artificiais

As redes neurais artificiais são um modelo de aprendizagem de máquina que se inspiram no funcionamento dos neurônios cerebrais para aprender e resolver problemas de alta complexidade.

Estas são compostas por neurônios matemáticos que interagem entre si dependendo da arquitetura em que essa rede se estrutura. 

image.png

Fig. Neurônio Matemático ou Artificial: Este neurônio possui na sua estrutura: as entradas (ou os estímulos), os pesos (a importância dada ao neurônio pelos estímulos), uma função de ativação e uma saída, sendo esta última, geralmente, um valor entre 0 e 1. Quando o valor é 0, dizemos que o neurônio não ativou, quando o valor é 1, dizemos que este foi ativado. 

O treinamento de um neurônio artificial consiste em regular os pesos dados às entradas, na tentativa de diminuir o erro, ou seja, fazer o neurônio ativar apenas quando tiver que ativar.

As redes neurais artificiais são muito versáteis e poderosas, e podemos imaginar isso a partir do seguinte exemplo: imagine que criemos uma camada de neurônios que em vez de fazer uma classificação, irá tratar os dados de entrada de modo a otimizá-los para uma futura classificação, podemos então acoplar a esta camada, uma outra que irá utilizar destes dados tratados para fazer realizar enfim a classificação. O que acontece é que esse processo, geralmente, oferece resultados poderosos em relação aos demais algoritmos de IA, pois você está utilizando uma camada de neurônios para otimizar os dados de entrada que serão utilizados por uma outra camada de neurônios que irão realizar a classificação. Inclusive, essas camadas inseridas na rede neural são chamadas de camadas ocultas, e a rede neural passa a ganhar o nome de rede neural profunda, e você pode inserir quantas camadas profundas quiser, o que claro, custará à você uma quantidade significativa de poder computacional para treinar e usar esta rede neural.
